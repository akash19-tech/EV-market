# Importing important libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# calling the dataset

```python
ev=pd.read_csv('EV2.csv')
```

```python
ev.head()
```

```python
ev.info()
```

```python
ev=ev.drop('review',axis=1)
```

# Implementing preprocesisng tasks

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()

ev['condition_final']=label_encoder.fit_transform(ev['Condition'])
ev['driven_final']=label_encoder.fit_transform(ev['driven'])
ev['model_name_final']=label_encoder.fit_transform(ev['model_name'])
```

```python
ev.head(50)
```

```python
ev.drop(['Condition','driven','model_name'],axis=1,inplace=True)
```

# checking the Relationship between model_name and Performance


0 : hyundai kona
1 : tata nexon ev
2 : tata tigor ev

```python
sns.lineplot(x='model_name_final',y='Performance',data=ev)
```

# Relationship between Perfromance and other columns

```python
columns_to_plot = ['model_name_final', 'driven_final', 'condition_final','Exterior','Comfort','Fuel Economy']

plt.figure(figsize=(12,12))
for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(len(columns_to_plot), 1, i)
    sns.lineplot(x=col, y='Performance', data=ev)
    plt.title(f'Performance vs {col}')
    plt.tight_layout()

plt.show()
```

# Relationship between Value for money and other columns

```python
columns_to_plot = ['model_name_final', 'driven_final', 'condition_final','Exterior','Comfort','Fuel Economy']

plt.figure(figsize=(12,12))
for i, col in enumerate(columns_to_plot, 1):
    plt.subplot(len(columns_to_plot), 1, i)
    sns.lineplot(x=col, y='Value for Money', data=ev)
    plt.title(f'Value for Money vs {col}')
    plt.tight_layout()

plt.show()
```

# This is the correlation matrix for all the features

```python
corr_matrix = ev.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Matrix")
plt.show()
```
